### 1.
A linear model is linear in its structure: y = mx+c, in which it assumes that a data is linearly proportional with a coefficient multiplying the x feature variable, hence assuming the predictor variable y to be increasing as x is multiplied by a scalar value proportionaly (linearly). In which an intercept is also derived to locate the starting point of the 'linear trend'. The model assumes that the data is either positively correlated, i.e. as x increases, so does y, or negatively correlated, i.e. as x decreases, so does y. 

### 2. 
The coefficient for the dummy variable represents the difference in the predicted outcome between that category and the reference category, meanwhile holding other variables constant. (in the case where the model includes an intercept) Hence, if the model includes the intercept, the intecept absorbs the 'baseline effect' for the reference category variable that is dropped during one hot encoding. In this case, each dummy coefficient shows how much higher or lower the expected outcome is compared to that reference category. By adding the intercept, together with the coefficient, the output of the model calculates the original data more accurately.

### 3.
Linear regression cannot be used for classification. The model itself can only predict and yield out continuous values as according to the structure of its equation. It cannot only output a 0 or 1 categorical value.

### 4.
When graphed out using scatter plots, if the datapoints are located above the diagonal line in of perfect correlation between the predicted and actual data, this is a sign of overfitting.

### 5. 
Multicollinearity occurs when two or more independent variables in a regression model are highly correlated with each other. This makes it hard for the model to isolate the individual effect of each predictor, thereby leading to errors in its prediction. The two stage least squares technique could potentially address this issue, by regressing the variable (that is highly correlated with another feature variable) using a different variable (i.e. Z, that is not correlated with the variable), the results of which would be the predicted values of this variable. Therefore, we could use the predicted value of the variable to model our actual prediction instead of using its original values. 

### 6.
By using a polynomial model, which is essentially adding powers based on the amount of features you have, could effectively capture non-linear relationships between the predictor and the independent variable (i.e. curvatures). 

### 7. 
The intercept is the predicted value of y when all feature variables are zero, or can be interpreted as the starting point of the predicted value's best fit line. The slope represents the change in predicted value of y for each unit increase in the feature variable x. The coefficient for dummy variables represents the difference in the predicted value of y between the hot-coded category and the reference category(the one you dropped from the start)

```python
#4. Split the sample 80/20 into a training and a test set. 
# Run a regression of `Price` on `Review Scores Rating` and `Neighbourhood `.
# What is the $R^2$ and RMSE on the test set? What is the coefficient on `Review Scores Rating`? 
# What is the most expensive kind of property you can rent?

from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, root_mean_squared_error

df_clean = dfh[['Price', 'Review Scores Rating', 'Neighbourhood']].dropna()

train_df, test_df = train_test_split(df_clean, test_size=0.2, random_state=112)

encoder = OneHotEncoder(drop='first')
encoder.fit(train_df[['Neighbourhood']])

X_train_category = encoder.transform(train_df[['Neighbourhood']])
X_test_category = encoder.transform(test_df[['Neighbourhood']])

review_num_train = train_df['Review Scores Rating'].values.reshape(-1, 1)
review_num_test = test_df['Review Scores Rating'].values.reshape(-1, 1)

X_train = np.hstack([review_num_train, X_train_category.toarray()])
X_test = np.hstack([review_num_test, X_test_category.toarray()])

y_train = train_df['Price'].values
y_test = test_df['Price'].values

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
r2 = r2_score(y_test, y_pred)
rmse = root_mean_squared_error(y_test, y_pred)

coef_review = model.coef_[0]

print(f"Test R^2: {r2:.4f}")
print(f"Test RMSE: {rmse:.4f}")
print(f"coefficient on review scores rating: {coef_review:.4f}")

most_expensive_neighborhood = (df_clean.groupby('Neighbourhood')['Price'].mean().idxmax())
print("Most expensive neighborhood on average:", most_expensive_neighborhood)
```

```python
# 5. Use the `PolynomialFeatures` function from `sklearn` to expand the set of numerical variables you're using in the regression. 
# As you increase the degree of the expansion, how do the $R^2$ and `RMSE` change? At what point does $R^2$ go negative on the test set? 
# For your best model with expanded features, what is the $R^2$ and `RMSE`? How does it compare to your best model from part 4?

from sklearn.preprocessing import PolynomialFeatures 
results = []
X_poly_train, X_poly_test, y_poly_train, y_poly_test = train_test_split(X_num, y, test_size=0.2, random_state=112)
for degree in range(0, 50):
    poly = PolynomialFeatures(degree=degree)
    X_poly_train_transform = poly.fit_transform(X_poly_train)
    X_poly_test_transform = poly.fit_transform(X_poly_test)

    model_poly = LinearRegression()
    model_poly.fit(X_poly_train_transform, y_poly_train)

    y_train_poly_pred = model_poly.predict(X_poly_train_transform)
    y_test_poly_pred = model_poly.predict(X_poly_test_transform)

    r2_train = r2_score(y_poly_train, y_train_poly_pred)
    r2_test = r2_score(y_poly_test, y_test_poly_pred)

    rmse_train = root_mean_squared_error(y_poly_train, y_train_poly_pred)
    rmse_test = root_mean_squared_error(y_poly_test, y_test_poly_pred)

    results.append((degree,r2_train, r2_test, rmse_train, rmse_test ))

df_results = pd.DataFrame(
    results,
    columns=['Degrees', 'Train R^2', 'Test R^2', 'Train RMSE', 'Test RMSE']
)

sns.lineplot(df_results, x='Degrees', y='Test R^2')
plt.axhline(0, color='red', linestyle='--')
plt.show()
point_of_which_negative = df_results[(df_results['Degrees'] != 0) & (df_results['Test R^2'] < 0)]['Degrees'].min()
print(f'The point of where Test R^2 goes negative is at degree: {point_of_which_negative}')
```

**Q10.** Let's look at a cousin of Linear Regression, called **kernel regression** or **local constant least squares** or **Nadaraya-Watson Estimator**.

We derived the OLS estimator for single linear regression by minimizing
$$
SSE(b_0, b_1) = \frac{1}{N}\sum_{i=1}^N (y_i - b_0 - b_1 x_i)^2
$$
with solution
$$
\hat{b}_0 = \bar{y} - \hat{b}_1 \bar{x}, \quad \hat{b}_1 = \dfrac{\frac{1}{N} \sum_{i=1}^N (x_i-\bar{x})(y_i - \bar{y})}{s_x^2}.
$$

When you step back and think about it, this is a bit weird: The algorithm is computing sample means, variances, and covariances, and using those to create a predictive model. The data themselves arguably vanish from the solution. This is elegant, this is strange.

Instead, let $k(z)$ be a kernel function, such as the Gaussian
$$
k(z) = \frac{1}{\sqrt{2\pi}} e^{-z^2/2}
$$
or uniform
$$
k(z) = \begin{cases}
1/2, & |z| \le \frac{1}{2}\\
0, & \text{otherwise.}
\end{cases}
$$
We'll instead minimize, or each predictor value $x$,
$$
SSE(\hat{y}(x)) = \dfrac{1}{N} \sum_{i=1}^N \left\lbrace y_i - \hat{y}(x)\right\rbrace^2 \dfrac{1}{h} k \left( \dfrac{x-x_i}{h} \right).
$$

1. Show that the optimal predictor is
$$
\hat{y}(x) = \dfrac{ \frac{1}{Nh} \sum_{i=1}^N  y_i k \left( \dfrac{x-x_i}{h} \right) }{\frac{1}{Nh} \sum_{i=1}^N k \left(  \dfrac{x-x_i}{h} \right)}
$$
This has many names, but let's call it the local constant least squares (LCLS) estimator, as opposed to ordinary least squares (OLS).
